```python
import numpy as np
import pandas as pd
import pickle
from scipy import stats
from matplotlib import pyplot as plt
```

```python
mice = ["mpfc14", "mpfc19", "mpfc20", "mpfc23", "mpfc24", "mpfc30"]
days = ["yd1", "yd2", "yd6"]
```

```python
x = []
for mi, mouse in enumerate(mice):
    xs = []
    for di, day in enumerate(days):
        xs.append(pd.read_csv(mouse + "/" + mouse + "_" + day + ".csv").values)
    x.append(xs)
```

```python
for mi, mouse in enumerate(mice):
    for di, day in enumerate(days):
        mu = np.array([x[mi][di][i*300:(i+1)*300, :].mean(axis=0) for i in range(x[mi][di].shape[0]//300)]).reshape(-1)
        sigma = np.array([x[mi][di][i*300:(i+1)*300, :].std(axis=0) for i in range(x[mi][di].shape[0]//300)]).reshape(-1)
        log_mu = np.log(mu[(mu > 0) & (sigma > 0)])
        log_sigma = np.log(sigma[(mu > 0) & (sigma > 0)])
        cov = np.cov(log_mu, log_sigma)
        a, b, r2, p, _ = stats.linregress(log_mu, log_sigma)
        xmin = log_mu.min()
        xmax = log_mu.max()
        plt.figure(figsize=(2, 2))
        plt.scatter(log_mu, log_sigma, s=2)
        plt.plot([xmin, xmax], [a * xmin + b, a * xmax + b], c="C1")
        ymin, ymax = plt.ylim()
        if ymin < -15:
            plt.ylim(-15, ymax)
        print(mouse, day)
        print("a=", a, "b=", b, "r2=", r2, "p=", p)
        print(np.corrcoef(log_mu, log_sigma)[0, 1])
        plt.tick_params(labelleft=False, labelbottom=False)
        plt.tight_layout()
        print(plt.xticks())
        print(plt.yticks())
        plt.savefig("paper/ext_fig1/" + mouse + "_" + day + ".png")
        plt.show()
```
