```python
import numpy as np
from matplotlib import pyplot as plt
import cupy as cp
import igpu
```

```python
for ri in range(1, 10):
    for si in range(10):
        x = igpu.open_csv_nonull("t3000/gen_n500_k5_l50_t3000_seed" + str(si + 40) + ".csv")
        w_save = None
        h_save = None
        k_save = None
        cost_min = np.inf
        for i in range(10):
            model = igpu.solve(x.T, 20, 50, z_th=x.min(), random_seed=i*100+ri*10+si, corr_max=ri*0.1, comp_rate=0.1)
            c = model.reconstruction_cost()
            print(ri*0.1, si, model.k, c)
            if c < cost_min:
                cost_min = c
                w_save = model.W.copy()
                h_save = model.H.copy()
                k_save = model.k
        np.savetxt("t3000/gen_n500_k5_l50_t3000_seed" + str(si + 40) + "_igpu_H.csv", cp.asnumpy(h_save).T, delimiter=",")
        for ki in range(k_save):
            np.savetxt("t3000/gen_n500_k5_l50_t3000_seed" + str(si + 40) + "_igpu_W" + str(ki) + ".csv", cp.asnumpy(w_save[:, ki, :]), delimiter=",")
```
